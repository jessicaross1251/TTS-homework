```python
class Complex(object):                             # We are going to create a class called Complex
    def __init__(self, real, imaginary):           # We need to initiate the class and say that what makes up itself
        self.real = real                           # is a real part, and an imaginary part.  
        self.imaginary = imaginary                 # We will define the method .real to obtain the first entry as the 
                                                   # real part, and we will define the method .imaginary to obtain the 
                                                   # second entry as the imaginary part.
    def __add__(self, no):
        return Complex(self.real + no.real, self.imaginary + no.imaginary)  # We will define the special function 
                                                   #__add__ to take the number self, and another number no and we will 
                                                   # add the real parts together, and add the imaginary parts together.
                                                   # We are returning this as an object in the Complex class.
    def __sub__(self, no):                                                 
        return Complex(self.real - no.real, self.imaginary - no.imaginary)  # We will define the special function
                                                   #__sub__ to take the number self, and the other number no and 
                                                   # subtract no's real part from self's real part, and subtract no's
                                                   # imaginary part from self's imaginary part.
                                                   # We again return the result as an object in the Complex class.
    def __mul__(self, no):
        return Complex(self.real * no.real - self.imaginary * no.imaginary, \  # We will define the special function
        self.imaginary * no.real + self.real * no.imaginary)                   #__mul__ to multiply the two complex
                                                   # numbers together.  To multiply complex numbers, you must treat
                                                   # it as if you are multiplying together two binomials.  Thus you
                                                   # must use the "FOIL" method, or multiply every part in the first
                                                   # complex number by every part in the second complex number.  It is
                                                   # worth noting that we are subtracting the product of the imaginary
                                                   # parts, because when we multiply i*i, the result is the real 
                                                   # number -1.  Thus this product will become a real number with the
                                                   # opposite sign.
                                                   # We will make this result a part of the Complex class.
    def __truediv__(self, no):
        return Complex((self.real*no.real + self.imaginary*no.imaginary)/(no.real**2 + no.imaginary**2),\
        (self.imaginary*no.real - self.real*no.imaginary)/(no.real**2 + no.imaginary**2))
                                                   # We will define the special function __truediv__ to divide the
                                                   # complex number self by the complex number no.  To divide complex
                                                   # numbers, we multiply the dividend and divisor by the conjugate
                                                   # of the divisor.  This will eliminate any imaginary parts in the
                                                   # divisor.  This will allow us to write the complex number in the
                                                   # form a + bi, where a and b are numbers, and i labels number b as
                                                   # imaginary.  Remember, when you multiply complex numbers, you must
                                                   # use the "FOIL" or distribution method.  Also recall that i*i=-1.
                                                   # We will return this result as a part of the Complex class.
        
    def mod(self):                                 # We will define a function called mod() that will perform modular
        return Complex((self.real**2 + self.imaginary**2)**0.5, 0)  # division on the complex number.  Modular
                                                                    # division for a complex number is the same formula
                                                                    # as finding the length of a vector.  The formula is
                                                                    # sqrt(a^2 + b^2).  Let us again return the answer
                                                                    # as a complex number.
    def __str__(self):
        if self.imaginary == 0:                     # Now, let us define a special function, __str__.  For this function
            result = "%.2f+0.00i" % (self.real)     # if the imaginary part of the self complex number is 0, then we
                                                    # want to write the result by giving the real part as a float 
                                                    # rounded to 2 decimal places plus 0.00i.
        elif self.real == 0:                        # Else, if the real part of the self complex number is 0, then we
            if self.imaginary >= 0:                 # want to first consider if that number is greater than or equal
                                                    # to 0.  If this is the case, we want to display the result as
                result = "0.00+%.2fi" % (self.imaginary)    # 0.00+ the imaginary part rounded to 2 decimal places with
                                                            # an i next to the imaginary part.
            else:                                   # Next we consider if the imaginary part of self is negative while.
                                                    # the real part is 0 .If it is, we will display
                result = "0.00-%.2fi" % (abs(self.imaginary))    # 0.00 - the absolute value of the imaginary part
                                                                 # rounded to 2 decimal places.
        elif self.imaginary > 0:                    # Next, we will consider what to do when the imaginary part is
            result = "%.2f+%.2fi" % (self.real, self.imaginary)   # positive while the real part is not 0.  We will 
                                                                  # display the real part rounded to 2 decimal places
                                                                  # + the imaginary part rounded to 2 decimal places
                                                                  # next to an i.
                                                    # Lastly, we will consider what to do when the imaginary part is
        else:                                       # negative while the real part is not 0.  We will desplay the real
            result = "%.2f-%.2fi" % (self.real, abs(self.imaginary)) # part rounded to 2 decimal places - the absolute
        return result                                                # value of the imaginary part of the number with 
                                                                     # an i next to it.
                                                    # After these options have been looked at, and the appropriate 
                                                    # action chosen, we will return the result.
                
    if __name__ == '__main__':              # Here we wish to tell Python that we only want to run the code inside the 
                                            # if statement when the program is run directly by the Python interpreter.
                                           
        c = map(float, input().split())     # If this case is satisfied, then we will split the first input into 2 
                                            # different float numbers and have variable c point to them.
        d = map(float, input().split())     # We will also split the second input into 2 float numbers and have
                                            # variable d point to these numbers.
        x = Complex(*c)                     # We will have x point to c, while making it a part of the Complex class.
                                            # "In the function, we should use an asterisk ( * ) before the parameter 
                                            # name to pass a variable number of arguments. Thus, we're sure that these
                                            # passed arguments make a tuple inside the function with the same name as
                                            # the parameter excluding * " -freeCodeCamp
        y = Complex(*d)                     # We will have y point to d, while making d a part of the complex class.
                                            # See the quote from freeCodeCamp above for comments on the *.
        print(*map(str, [x+y, x-y, x*y, x/y, x.mod(), y.mod()]), sep = '\n')  # Finally, we will print a mapping of
                                                                              # the str function using the special 
                                                                              # functions we created, +, -, *, /, and 
                                                                              # mod. We will separate these by starting
                                                                              # a new line after each computation.
```
